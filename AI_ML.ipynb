```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler

data = {
    "Age": [18, 20, np.nan, 22, 21],
    "Gender": ["Male", "Female", "Female", "Male", "Male"],
    "Score": [65, 75, 80, np.nan, 90],
    "Passed": ["No", "Yes", "Yes", "Yes", "Yes"]
}

df = pd.DataFrame(data)

# Handle missing values
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Score"].fillna(df["Score"].mean(), inplace=True)

print("Before Encoding:\n", df)

# Encoding categorical data
gender_encoder = LabelEncoder()
df["Gender"] = gender_encoder.fit_transform(df["Gender"])

passed_encoder = LabelEncoder()
df["Passed"] = passed_encoder.fit_transform(df["Passed"])

print("\nAfter Encoding:\n", df)

# Scaling numeric columns
scaler = StandardScaler()
df[["Age", "Score"]] = scaler.fit_transform(df[["Age", "Score"]])

print("\nAfter Scaling:\n", df)




```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, accuracy_score

# Linear Regression Example (Predicting the house price)
x = np.array([[50], [60], [70], [80], [90]]) #square meters
y = np.array([150, 180, 210, 240, 270])

model_lr = LinearRegression()
model_lr.fit(x,y)

predicted_price = model_lr.predict([[75]])
# print(predicted_price[0])


# Logistic Regression Example( Predict pass/fail)
data = {
    "Hours_Studied": [1,2,3,4,5,6],
    "Passed": [0,0,0,1,1,1] 
}

df = pd.DataFrame(data)
x_log = df[["Hours_Studied"]]
y_log = df[["Passed"]]

model_log = LogisticRegression()
model_log.fit(x_log, y_log)

# Predicting if a student passes after studing for 3.5 hours
predicted_result = model_log.predict([[3.5]])

print(predicted_result)
print("Predicted result for 3.5 hours:", "Pass" if predicted_result[0] == 1 else "fail")


# Evaluating the models
y_pred_lr = model_lr.predict(x)
mse = mean_squared_error(y, y_pred_lr)
# print(y_pred_lr)
print("Linear Regression Mean Squared Error:", mse)

y_pred_log = model_log.predict(x_log)
accuracy = accuracy_score(y_log, y_pred_log)
print("Logistic Regression Accuracy: ", accuracy)
```
