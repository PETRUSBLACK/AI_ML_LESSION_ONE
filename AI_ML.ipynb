```python
import numpy as np

#1. Creating arrays
arr1 = np.array([10, 20, 30, 40, 50])
arr2 = np.array([
    [1,2,3],
    [4,5,6]
])

print("iD Array", arr1)
print("2D Array: \n", arr2)

#2. Shape and size
print("Shape of arr2", arr2.shape)
print("Size of arr2", arr2.size)

#3. Array operations
print("Addition", arr1 + 5)
print("Multiplication", arr1 + 2)

#4. Mathematical functions
print("Mean:", np.mean(arr1))
print("Max:", np.max(arr1))
print("Min:", np.min(arr1))

#5. Indexing and slicing
print("First element:", arr1[0])
print("Slice", arr1[1:4])

#6. Reshaping arrays
reshaped = arr1.reshape(5,1)
print("reshaped array:", reshaped)

#7. Random numbers
random_arr = np.random.randint(1, 100, 5)

#8. Matrix multiplication
A = np.array([[1,2], [3,4]])
B = np.array([[5,6], [7,8]])
C = np.dot(A,B)
print("Matrix multiplication", C)
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, accuracy_score

#Linear Regression. Example (Predicts house price)
x = np.array([[50], [60], [70], [80], [90]])
y = np.array([150, 180, 210, 240, 270])

model_lr = LinearRegression()
model_lr.fit(x,y)

predicted_price = model_lr.predict([[75]])
print("Predicted price for 75 sq.m:", predicted_price[0])
```
